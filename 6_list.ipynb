# List 
A list is ordered , mutabl (changeable) collection that can store multiple values in a single variable . List is use square braket [] and values is separate with comma (,).
- List ordered hoti hai
- List duplicate values allow karti hai
- List mutable hoti hai (change ho sakti hai)
- List any data type store kar sakti hai
```python
list_name = [item,item1,item2]
```


- list_name → variable ka naam
- [] → list define karne ke liye
- item1, item2 → elements (any data type allowed)

````python
#Empty list 
my_list = []
````

````python
data = [1,3,4,"Python",3.4,True]
print(data)
````

Accessing Elements (Indexing)
Indexing always start from zero 


````python
#make a list of fruits and access items with indexinig 
fruits = ["Apple","Banana","Orange"]
print(fruits[0])
print(fruits[1])
#change the data at index 0 
fruits[0] = "Pineapple"
print(fruits)

#add new fruit using append function at last of list
fruits.append("Apple")
print(fruits)
````

Loop through list

````python
#use For Loop to print element
marks = [32,22,12,54]
for m in marks:
    print(m)
````

Taking Input from User

````python
#Single Input
items = input("Enter an item")

my_list=[items]
print(my_list)
````

````python
#Multiple Inputs
item = input("Enter items Speatated by space :").split()
print(item)
````

````python
#Multiple Integer Inputs 
numbers = list(map(int,input("Enter numbers :").split()))
print(numbers)
````

````python
#Taking input using loop 
n = int(input("how many elements"))
list1 = []
for i in range(n):
    value = input("Enter values")
    list1.append(value)

print(list1)
````
