# 3. Basic Operators
Operators perform operations on values. Python supports arithmetic, comparison,
and logical operators.


Arithmetic Operators

```python
a , b = 10 , 3 
print(a+b) #addition
print(a-b) #subtraction
print(a*b) #multiplication
print(a/b) #division
print(a//b) #floor division
print(a%b) #reminder
```

Comparison Operators

```python
#comapare values and return True/False
#a , b = 10 , 3 
print(a>b)
print(a<b)
print(a==b)
print(a!=b)
```

Logical Operators
1. And (Multiplication)
2. OR (Addition)
3. not (inverse)

```python
print(True and False) # 1*0=0 (false)
print(True or False) # 1+0=1 (True)
print(not True) # False
```
