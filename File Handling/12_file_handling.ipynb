# File Handling in Python

File handling in Python involves interacting with files on your computer to **read** data from them or **write** data to them. <br>
Python provide some built in functions : 
- creating 
- opening 
- reading 
- writing 
- closing 

**Opening a File in Python**
- To perform any file operation, the first step is to open
the file.
- Python's built-in open() function is used to open files
in various modes, such as reading, writing, and
appending.
```Python 
file = open("filename","mode")

Where , **filename** is the name of the file to open and **mode** in which file is opened eg reading 'r' , writing 'w' and 'a' for appending.

````python
fo = open("foo.txt" , "wb")
print("name of the file ",fo.name)
print("closed or not ",fo.closed)
print("opening mode ",fo.mode)
````

````python
fo = open("foo.txt" , "a")
````

````python
with open("foo.txt" , "w") as fo:
    fo.write("Python is a great language.\nYeah its great!!\n")
````

````python
lines = ["life is beautiful \n", "I love programming\n", "Let's code more\n"]

with open("foo.txt" , 'w') as fo:
    fo.writelines(lines)
````

````python
lines = ["life is beautiful \n", "I love programming\n", "Let's code more\n"]
with open("foo.txt" , 'a') as fo:
    fo.writelines(lines)
````

````python
with open("foo.txt" , "r") as fo:
    str = fo.read()
    print(str)
````

````python
with open("foo.txt" , 'a') as fo:
    fo.write("I love Python programming.\n")
````
