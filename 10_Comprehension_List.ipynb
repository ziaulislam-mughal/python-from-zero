# Comprehension List
A **Comprehension list** is a concise way to create a list in python . Instead of writing a loop and appending elements one by one , you can generate the list in a single line. 
<br>
It make your code shorter , readable and faster.

**Syntax** <br>
```python
[expression for item in iterable if condition]


- **expression** → The value to store in the new list (can be the item - itself or a calculation on it)

- **item** → The variable representing each element in the iterable

- **iterable** → Any iterable object (like a list, range, string, etc.)

- **condition** (optional) → A filter to include elements that satisfy this condition

**Example 1:** Basic List Comprehension
Create a list of squares of numbers from 1 to 5:

````python
square = [x**2 for x in range(1,6)]
print(square)
````

**Example 2:** With Condition,
Create a list of even numbers from 1 to 10:

````python
even_number = [x for x in range(1,11) if x%2==0 ]
print(even_number)
````
