# 1. Variables & Basic Types

Variables store data values. Python automatically determines the data type based on
the value assigned.

```python
name = "Python" #Output : string
age = 30 #Output : integar
price = 23.43 #Output : float
is_active = True #Output : boolean

print(name)
print(age)
print(price)
print(is_active)
```

# How we check datatype of data ?
Key Types: str (text), int (whole numbers), float (decimals), bool (True/False)

```python
print(type(name))
```

# 2. Strings & f-strings
Strings are sequences of characters. f-strings allow embedding variables directly into
text using curly braces.

String Operations

```python
text = "Hellow"
print(text.upper()) #Output : HELLOW
print(text.lower()) #Output : hellow
print(text.replace("Hellow" ,"Python")) #Output : hello replace with Python
```

f-strings (Formatted String Literals)

```python
#f-string allows embedding variables 
name = "Alice"
score = 56
message = f"Hellow {name} , your score is {score}"
print(message)
```

```python
#Multiple line string 

info = f"""
Name  : {name}
Score : {score}
Grade : B
"""
print(info)
```
