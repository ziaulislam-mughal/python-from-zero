# Tuple 
A tuple in python is an ordered , immutable (cannot be changed) collection of elements. It is used when you want to fix data that should not be modified accidentally.

```Python 
tuple_name = (item1, item2, item3)

**Important Rules**

- Tuples are written using round brackets ( )

- They allow duplicate values

- They can store multiple data types

- Once created, elements cannot be changed, added, or removed

````python
#Basic tuple
my_tuple = (1,2,3)
#Tuple with different datatype
mixed_tuple = (1,'python',3.5,True)
#Tuple packing 
packed_tuple = 12,2,43
# Single Tuple
single_tuple = (3,)
# Empty Tuple
empty_tuple = ()
````

**Acessing Tuple**

````python
print(my_tuple[0]) #Access first item in my_tuple
#use negative indexing to access last element 
print(my_tuple[-1])
````

**Tuple Slicing**

````python
numbers = (10, 20, 30, 40, 50)

print(numbers[1:4])   # (20, 30, 40)
print(numbers[:3])    # (10, 20, 30)
print(numbers[::2])   # (10, 30, 50)
````

‚ùå Tuple is Immutable (Cannot Change)

````python
data = (1, 2, 3)

# This will cause an error
# data[0] = 10
````

Tuple Operation

````python
a = (1, 2)
b = (3, 4)

# Concatenation
c = a + b
print(c)   # (1, 2, 3, 4)

# Repetition
print(a * 3)   # (1, 2, 1, 2, 1, 2)

# Membership
print(2 in a)   # True
````

**Getting Input from User in Tuple** 
---
üîπ Single Input (Convert to Tuple)

````python
value = input("Enter a value: ")
my_tuple = (value,)

print(my_tuple)
print(type(my_tuple))
````

Multiple Inputs (Using `split()`)

````python
user_input = input("Enter multiple values separated by space: ")
my_tuple = tuple(user_input.split())

print(my_tuple)
````

üß† Summary

- Tuple is ordered & immutable
- Defined using ()

- Ideal for fixed data
- Supports indexing, slicing, looping
- Input can be taken using input(), split(), and map()
