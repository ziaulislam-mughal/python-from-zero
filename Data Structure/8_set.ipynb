# Python Set 

This notebook explains Python Sets with syntax, examples, and AI use cases.



## What is a Set?
A set is an unordered collection of unique elements. It removes duplicates automatically and is widely used in AI and data processing.

```python
numbers = {1, 2, 3, 4, 5}
print(numbers)

duplicates = {1, 2, 2, 3, 3, 4}
print(duplicates)
```

## Set Syntax
Sets can be created using curly braces or the set() constructor.

Note: Empty set must be created using set().

```python
fruits = {'apple', 'banana', 'mango'}
print(fruits)

numbers_list = [1, 2, 3, 4, 4, 5]
numbers_set = set(numbers_list)
print(numbers_set)

empty_set = set()
print(type(empty_set))
```

## Taking User Input in Set

```python
value = input('Enter a value: ')
user_set = {value}
print(user_set)
```

```python
values = input('Enter values separated by space: ').split()
user_set = set(values)
print(user_set)
```

## Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print('Union:', A | B)
print('Intersection:', A & B)
print('Difference A-B:', A - B)
print('Symmetric Difference:', A ^ B)
```

## Special Use of Set in AI Engineering

```python
features = ['age', 'salary', 'age', 'experience', 'salary']
unique_features = set(features)
print(unique_features)

sentence = 'ai is the future and ai is powerful'
vocabulary = set(sentence.split())
print(vocabulary)

dataset1 = {'height', 'weight', 'age'}
dataset2 = {'age', 'salary', 'experience'}
print(dataset1 & dataset2)
```
