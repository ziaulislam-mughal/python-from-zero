# Exception Handling in Python

In Python, **`try`** and **`except`** are used for **exception handling**, which means handling errors gracefully instead of letting the program crash.


**Syntax**
```python 
try:
    #code that may raise an exception
    risky_code
except SomeException:
    #code that runs if an exception occurs
    handdle_exception 

## How `try` and `except` Work

- **`try:`** → You put the code that might fail here.  

- **`except SomeException:`** → You catch the specific error you expect and handle it.  

- You can catch **multiple exceptions** or use a **generic `except:`** for any error.


Example:

````python
try: 
    num1 = int(input("Enter a number:"))
    num2 = int(input("Enter another number:"))
    result = num1 / num2 
    print("The result of division is:", result)
except ZeroDivisionError:
    print("Error : You cannot divide by zero.")
except ValueError:
    print("Error : Invalid input. Please enter numeric values only.")
````

## What Happens in Different Cases

- If the user enters `0` for `num2`, a **`ZeroDivisionError`** is caught.  

- If the user enters a letter instead of a number, a **`ValueError`** is caught.

# Exception Handling in Python

In Python, **`try`** and **`except`** are used for **exception handling**, which means handling errors gracefully instead of letting the program crash.

## How `try` and `except` Work

- **`try:`** → You put the code that might fail here.  
- **`except SomeException:`** → You catch the specific error you expect and handle it.  
- You can catch **multiple exceptions** or use a **generic `except:`** for any error.

## What Happens in Different Cases

- If the user enters `0` for `num2`, a **`ZeroDivisionError`** is caught.  
- If the user enters a letter instead of a number, a **`ValueError`** is caught.

## Example: Catching Any Exception




````python
try:
    num = int(input("Enter number"))
    print(num)
except ValueError:
    print("Invalid input")
else:
    print("no error occur , input is valid ")
finally:
    print("Execution completed. and it is finally block")
````

## `else` and `finally` in Exception Handling

- **`else:`** → Runs if **no exception occurs** in the `try` block.  

- **`finally:`** → **Always runs**, whether there was an exception or not.  
  This is useful for **cleanup tasks**, like closing files or releasing resources.


## ✅ Summary in Simple Words

- **`try`** → Try this code, it might fail.  
- **`except`** → If it fails, do this instead of crashing.  
- **`else`** → Runs if everything is fine.  
- **`finally`** → Runs no matter what, used for cleanup.
