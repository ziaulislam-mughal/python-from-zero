# üß† Object-Oriented Programming (OOP) in Python

This notebook covers **Classes, Methods, Constructors, Dunder Methods, and 4 Pillars of OOP** with senior-level examples.

## 1Ô∏è‚É£ Class and Object

```

class Student:
    name = "Zia"
    age = 22

s1 = Student()
print(s1.name)
print(s1.age)
```

## 2Ô∏è‚É£ Methods

```

class Student:
    def study(self):
        print("Student is studying")

s1 = Student()
s1.study()
```

## 3Ô∏è‚É£ Constructor (__init__)

```

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Zia", 22)
print(s1.name, s1.age)
```

## 4Ô∏è‚É£ Dunder Method Example

```

class Book:
    def __init__(self, pages):
        self.pages = pages

    def __len__(self):
        return self.pages

b = Book(300)
print(len(b))
```

# üîë Four Pillars of OOP

## 1Ô∏è‚É£ Abstraction

```

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

c = Car()
c.start()
```

## 2Ô∏è‚É£ Encapsulation

```

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        self.__balance += amount

acc = BankAccount(1000)
print(acc.get_balance())
```

## 3Ô∏è‚É£ Inheritance

```

class Person:
    def speak(self):
        print("I can speak")

class Student(Person):
    def study(self):
        print("I can study")

s = Student()
s.speak()
s.study()
```

## 4Ô∏è‚É£ Polymorphism

```

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]
for a in animals:
    a.sound()
```
